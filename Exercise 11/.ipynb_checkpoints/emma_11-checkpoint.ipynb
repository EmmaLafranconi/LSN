### La Regressione nel Machine Learning

Consideriamo dati etichettati $(x, y)$, generati campionando dall'equazione

$$ y_i = f(x_i) + \eta_i $$

dove $f(x_i)$ è una funzione fissa ma possibilmente sconosciuta e $\eta_i$ è una variabile di rumore gaussiano non correlato tale che

$$ \langle \eta_i \rangle = 0 \quad \text{e} \quad \langle \eta_i \eta_j \rangle = \delta_{ij} \sigma $$

Per fare previsioni, considereremo una rete neurale (NN) che dipende dai suoi parametri, *pesi* e *bias*. Le funzioni che la NN può modellare rappresentano la *classe di modelli* che stiamo usando per cercare di modellare i dati e fare previsioni. Per apprendere i parametri della NN, addestreremo i nostri modelli minimizzando la funzione di costo su un set di *dati di addestramento* e poi testeremo l'efficacia della NN su un set di dati diverso, il set di *dati di validazione*. La ragione per cui dobbiamo dividere i nostri dati in un set di addestramento e un set di test è che lo scopo del *machine learning* è fare previsioni accurate su nuovi dati che non abbiamo ancora visto. Una misura comune delle prestazioni predittive del nostro algoritmo è confrontare le previsioni, $\{y_j^{\text{pred}}\}$, con i valori reali $\{y_j\}$. 

### Exercise 11.1

In questo esercizio viene chiesto di eseguire un fit lineare su dati rumorosi con una rete neurale (NN) della funzione $f(x)=2x+1$.

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation
from tensorflow.keras import backend as K
from tensorflow.keras.utils import get_custom_objects
```

```python
# generate training inputs
np.random.seed(0)
x_train = np.random.uniform(-1, 1, 1000)
x_valid = np.random.uniform(-1, 1, 100)
x_valid.sort()

# target parameters of f(x) = m*x + b
m = 2 # slope
b = 1 # intersect
y_target = m * x_valid + b # ideal (target) linear function

sigma = 0.1 # noise standard deviation
y_train = np.random.normal(m * x_train + b, sigma) # actual measures from which we want to guess regression parameters
y_valid = np.random.normal(m * x_valid + b, sigma)

# plot validation and target dataset
plt.plot(x_valid, y_target, label='target')
plt.scatter(x_valid, y_valid, color='r', label='validation data', marker='.')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()
```

Nota che in questo caso, poichè i neuroni funzionano esattamente come funzioni lineari, è sufficiente un singolo neurone.

```python
# compose the NN model
model = Sequential()
model.add(Dense(1, input_shape=(1,)))

# compile the model choosing optimizer, loss and metrics objects
model.compile(optimizer='sgd', loss='mse', metrics=['mse'])

# get a summary of our composed model
model.summary()
```

```python
# fit the model using training dataset
# report training progress against validation data
history = model.fit(x=x_train, y=y_train, 
          batch_size=32, epochs=80,
          shuffle=True, # a good idea is to shuffle input before at each epoch
          validation_data=(x_valid, y_valid))
```

Guardando i pesi e i *bias* possiamo vedere se il fit lineare è riuscito: $ w_1$ rappresenta il coefficiente angolare, $b$ l'intercetta.

```python
# return weights and biases
model.get_weights()
```

```python
# evaluate model
score = model.evaluate(x_valid, y_valid, batch_size=32, verbose=1)

# print performance
print()
print('Test loss:', score[0])
print('Test accuracy:', score[1])
```

```python
# evaluate model with the exact curve
score = model.evaluate(x_valid, y_target, batch_size=32, verbose=1)

# print performance
print()
print('Test loss:', score[0])
print('Test accuracy:', score[1])
```

```python
# look into training history
# plot training & validation loss values
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('Model loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['Train', 'Test'], loc='best')
plt.show()
```

Osservando il grafico della funzione di costo durante l'addestramento, si può notare che aumentando il numero di epoche, il numero di dati di input e aggiungendo rumore, il modello apprende più in fretta.

```python
x_predicted = np.random.uniform(-1, 1, 100)
y_predicted = model.predict(x_predicted)
plt.scatter(x_predicted, y_predicted, color='r', label='predicted data', marker='.')
plt.plot(x_valid, y_target, label='target')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()
```

Dal grafico si può osservare che il modello, dopo aver regolato alcuni iperparametri, ha una capacità predittiva molto accurata. 

### Exercise 11.2

In questo esercizio viene chiesto di addestrare una rete neurale per ottenere un fit ragionevole del seguente polinomio di terzo grado: 

$$
f(x)=4-3x-2x^2+3x^3
$$

per $x \in [-1,1]$. L'obiettivo è anche sviluppare un'intuizione sul motivo per cui la previsione è difficile.

```python
def f(x):
    return 4 - 3*x - 2*x**2 + 3*x**3

# generate training inputs
np.random.seed(0)
x_train = np.random.uniform(-1, 1, 5000)
x_valid = np.random.uniform(-1, 1, 500)
x_valid.sort()
y_target = f(x_valid) # ideal (target) polynomial

sigma = 0.2 # noise standard deviation
y_train = np.random.normal(f(x_train), sigma) # actual measures from which we want to guess regression parameters
y_valid = np.random.normal(f(x_valid), sigma)

# plot validation and target dataset
plt.plot(x_valid, y_target, label='target')
plt.scatter(x_valid, y_valid, color='r', label='validation data', marker='.')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()
```

A causa del comportamento non lineare della funzione, la rete neurale non può più essere costituita da un solo neurone senza attivazione. Dato che un generico polinomio di terzo grado dipende da 4 parametri, una rete neurale con un solo *hidden layer* di 2 neuroni dovrebbe essere sufficiente. Per esplorare come la regressione dipende dalla struttura della rete neurale, sono stati creati due modelli con diverse funzioni di attivazione, ottimizzatori e funzioni di costo.

```python
from keras.optimizers import Adam

def create_model_1():
    # compose the NN model
    model = Sequential()
    model.add(Dense(2, input_shape=(1,), activation='sigmoid'))
    model.add(Dense(1))
    # compile the model choosing optimizer, loss and metrics objects
    model.compile(optimizer='sgd', loss='mse', metrics=['mse'])
    return model

def create_model_2():
    # compose the NN model
    model = Sequential()
    model.add(Dense(2, input_shape=(1,), activation='tanh'))
    model.add(Dense(1))
    # compile the model choosing optimizer, loss and metrics objects
    model.compile(optimizer=Adam(lr=0.01), loss='msle', metrics=['mse'])
    return model
```

```python
# create the deep neural net
model_1 = create_model_1()
model_2 = create_model_2()

# fit the model using training dataset
# report training progress against validation data

print()
print('Model 1:')
history_1 = model_1.fit(x=x_train, y=y_train, 
          batch_size=32, epochs=150,
          shuffle=True, # a good idea is to shuffle input before at each epoch
          validation_data=(x_valid, y_valid))

print()
print('Model 2:')
history_2 = model_2.fit(x=x_train, y=y_train, 
          batch_size=32, epochs=100,
          shuffle=True, # a good idea is to shuffle input before at each epoch
          validation_data=(x_valid, y_valid))
```

```python
# evaluate model
score_1 = model_1.evaluate(x_valid, y_valid, batch_size=32, verbose=1)
score_2 = model_2.evaluate(x_valid, y_valid, batch_size=32, verbose=1)

# print performance
print()
print('Model 1:')
print('Test loss:', score_1[0])
print('Test accuracy:', score_1[1])

print()
print('Model 2:')
print('Test loss:', score_2[0])
print('Test accuracy:', score_2[1])
```

```python
# evaluate model with the exact curve
score_1 = model_1.evaluate(x_valid, y_target, batch_size=32, verbose=1)
score_2 = model_2.evaluate(x_valid, y_target, batch_size=32, verbose=1)

# print performance
print()
print('Model 1:')
print('Test loss:', score_1[0])
print('Test accuracy:', score_1[1])

print()
print('Model 2:')
print('Test loss:', score_2[0])
print('Test accuracy:', score_2[1])
```

```python
# look into training history
# plot training & validation loss values
def plot_training_history(history, title):
    plt.plot(history.history['loss'])
    plt.plot(history.history['val_loss'])
    plt.title(title)
    plt.ylabel('Loss')
    plt.xlabel('Epoch')
    plt.legend(['Train', 'Test'], loc='best')
    plt.show()

# Plot for each model
plot_training_history(history_1, 'Model 1')
plot_training_history(history_2, 'Model 2')
```

Dai grafici si può osservare che il Modello 2 converge più rapidameente rispetto al Modello 1.

```python
x_predicted = np.random.uniform(-1, 1, 100)
y_predicted_1 = model_1.predict(x_predicted)
y_predicted_2 = model_2.predict(x_predicted)
plt.scatter(x_predicted, y_predicted_1, color='green', label='predicted data with model 1', marker='.')
plt.scatter(x_predicted, y_predicted_2, color='r', label='predicted data with model 2', marker='.')
plt.plot(x_valid, y_target, label='target')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()
```

Considerando 150 epoche per l'addestramento, il Modello 1 non predice bene il nuovo set di dati generati, mentre il Modello 2 ha una buona capacità predittiva anche con 100 epoche. Possiamo concludere che a parità di complessità del modello, una rete del tipo Modello 2 è più adatta a fittare un polinomio di terzo grado.

Osserviamo ora cosa accade se proviamo a estrapolare punti al di fuori del range su cui è stato addestrato il Modello 2.

```python
x = np.arange(-1.5, 1.5, 0.1)
x_predicted = np.random.uniform(-1.5, 1.5, 100)
y_predicted = model_2.predict(x_predicted)
plt.scatter(x_predicted, y_predicted, color='r', label='predicted data', marker='.')
plt.plot(x, f(x), label='target')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()
```

Dal grafico si può osservare che il modello si comporta bene sui punti nell'intervallo in cui è stato addestrato, ma performa decisamente male al di fuori.

### Exercise 11.3

In questo esercizio viene chiesto di estendere il modello per fittare la funzione trigonometrica bidimensionale 

$$f(x, y) = \sin(x^2 + y^2)$$

nell'intervallo $x \in [-\frac{3}{2}, \frac{3}{2}]$ e $y \in [-\frac{3}{2}, \frac{3}{2}]$.

```python
def F(x, y):
    return np.sin(x**2 + y**2)

# generate training inputs
np.random.seed(0)
data_train = np.random.uniform(-1.5, 1.5, size=(10000, 2))
data_valid = np.random.uniform(-1.5, 1.5, size=(1000, 2))
z_target = F(data_valid[:,0], data_valid[:,1]) # ideal (target) function

sigma = 0.2 # noise standard deviation
z_train = np.random.normal(F(data_train[:,0], data_train[:,1]), sigma)
z_valid = np.random.normal(z_target, sigma)

# plot validation and target dataset
fig = plt.figure(figsize=(12,10))
ax = fig.add_subplot(projection='3d')
x = np.arange(-1.5, 1.5, 0.1)
y = np.arange(-1.5, 1.5, 0.1)
X, Y = np.meshgrid(x, y)
ax.plot_wireframe(X, Y, F(X, Y), rstride=2, cstride=2, label='target', alpha = 0.5)
ax.scatter(data_valid[:,0], data_valid[:,1], z_valid, color='r', label='validation data', marker='.')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.view_init(elev=15, azim=25)
ax.legend()
plt.show()
```

Data la complessità del problema, è stata costruita una reta neurale più profonda, con 2 *hidden layer* di 32 neuroni ciascuno. In questo caso, si è scelto di utilizzare la funzione di attivazione 'ReLU' che non satura per valori di input elevati.

```python
# compose the NN model
model = Sequential()
model.add(Dense(32, input_shape=(2,), activation='relu'))
model.add(Dense(32, activation='tanh'))
model.add(Dense(1))

# compile the model choosing optimizer, loss and metrics objects
model.compile(optimizer='nadam', loss='logcosh', metrics=['mse'])

# get a summary of our composed model
model.summary()
```

```python
# fit the model using training dataset
# report training progress against validation data
history = model.fit(x=data_train, y=z_train, 
          batch_size=32, epochs=100,
          shuffle=True, # a good idea is to shuffle input before at each epoch
          validation_data=(data_valid, z_valid))
```

```python
# evaluate model
score = model.evaluate(data_valid, z_valid, batch_size=32, verbose=1)

# print performance
print()
print('Test loss:', score[0])
print('Test accuracy:', score[1])
```

```python
# evaluate model with the exact curve
score = model.evaluate(data_valid, z_target, batch_size=32, verbose=1)

# print performance
print()
print('Test loss:', score[0])
print('Test accuracy:', score[1])
```

```python
# look into training history
# plot training & validation loss values
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('Model loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['Train', 'Test'], loc='best')
plt.show()
```

```python
data_predicted = np.random.uniform(-1.5, 1.5, size=(500, 2))
z_predicted = model.predict(data_predicted)
fig = plt.figure(figsize=(12,10))
ax = fig.add_subplot(projection='3d')
ax.plot_wireframe(X, Y, F(X, Y), rstride=2, cstride=2, label='target', alpha = 0.5)
ax.scatter(data_predicted[:,0], data_predicted[:,1], z_predicted, color='r', label='predicted data', marker='.')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.view_init(elev=15, azim=25)
ax.legend()
plt.show()
```

Osserviamo un buon fit sul dominio di allenamento.
