### Exercise 12.1

In questo esercizio viene chiesto di utilizzare modelli di reti neurali profonde (DNN) per riconoscere e distinguere tra le dieci cifre scritte a mano (0-9). Il dataset MNIST consiste in 70000 cifre scritte a mano, ognuna delle quali è rappresentata da un'immagine quadrata suddivisa in una griglia di pixel 28x28. Ogni pixel può assumere 256 gradazioni di colore grigio, interpolando tra il bianco e il nero, quindi ogni dato può assumere un valore nel set {0, 1, ..., 255}. Poiché ci sono 10 categorie nel problema, corrispondenti alle dieci cifre, questo problema rappresenta un compito generico di classificazione.

```python
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow import keras
from keras.datasets import mnist
from keras.models import Sequential
from keras.layers import Dense, Dropout
import os

seed=0
np.random.seed(seed) # fix random seed
tf.random.set_seed(seed)
```

#### Load and Process the Data

```python
# input image dimensions
img_rows, img_cols = 28, 28 # number of pixels 
# output
num_classes = 10 # 10 digits

# the data, split between train and test sets
(X_train, Y_train), (X_test, Y_test) = mnist.load_data()

print('X_train shape:', X_train.shape)
print('Y_train shape:', Y_train.shape)
```

#### Reshape data and convert labels to be used with categorical cross entropy

```python
# reshape data, it could depend on Keras backend
X_train = X_train.reshape(X_train.shape[0], img_rows*img_cols)
X_test = X_test.reshape(X_test.shape[0], img_rows*img_cols)
print('X_train shape:', X_train.shape)
print('X_test shape:', X_test.shape)
print()

# cast to floats
X_train = X_train.astype('float32')
X_test = X_test.astype('float32')

# rescale data in interval [0,1]
X_train /= 255
X_test /= 255

# convert class vectors to binary class matrices, e.g. for use with categorical_crossentropy
Y_train = keras.utils.to_categorical(Y_train, num_classes)
Y_test = keras.utils.to_categorical(Y_test, num_classes)
print('X_train shape:', X_train.shape)
print('Y_train shape:', Y_train.shape)
```

#### Define the Neural Net and its Architecture

Nell'architettura della rete neurale è stato applicato il *dropout*. Il *dropout* consiste nel settare casualmente una frazione delle unità di *input* a 0 durante ciascun aggiornamento nel tempo di addestramento, tecnica utilizzata per prevenire l'*overfitting* nella rete neurale. Infatti, il *dropout* migliora la generalizzazione, irrobustendo così il modello, perché costringe uno strato a imparare con neuroni diversi. Durante la fase di test, il *dropout* è disattivato. Di conseguenza, l'accuratezza del test migliorerà rispetto a quella di addestramento.

```python
def create_DNN():
    # instantiate model
    model = Sequential()
    # add a dense all-to-all relu layer
    model.add(Dense(400, input_shape=(img_rows*img_cols,), activation='relu'))
    # add a dense all-to-all relu layer
    model.add(Dense(100, activation='relu'))
    # apply dropout with rate 0.5
    model.add(Dropout(0.5))
    # soft-max layer
    model.add(Dense(num_classes, activation='softmax')) # normalizes the vector into a probability distribution.
    
    return model

print('Model architecture created successfully!')
```

#### Choose the Optimizer and the Cost Function

Per valutare la predizione di diversi modelli sui dati, sono stati creati tre modelli con ottimizzatori differenti. 

```python
from keras.optimizers import SGD, Adam, RMSprop, Adagrad, Adadelta, Adam, Adamax, Nadam

def compile_model_SGD():
    # create the model
    model=create_DNN()
    # compile the model
    model.compile(loss=keras.losses.categorical_crossentropy,
                  optimizer=SGD(),
                  metrics=['acc'])
    return model

def compile_model_Adam():
    # create the model
    model=create_DNN()
    # compile the model
    model.compile(loss=keras.losses.categorical_crossentropy,
                  optimizer=Adam(),
                  metrics=['acc'])
    return model

def compile_model_RMSprop():
    # create the model
    model=create_DNN()
    # compile the model
    model.compile(loss=keras.losses.categorical_crossentropy,
                  optimizer=RMSprop(),
                  metrics=['acc'])
    return model

print('Models compiled successfully and ready to be trained.')
```

#### Train the model

```python
# training parameters
batch_size = 32
epochs = 50 

# create the deep neural net
model_DNN_SGD = compile_model_SGD()
model_DNN_Adam = compile_model_Adam()
model_DNN_RMSprop =compile_model_RMSprop()

# train DNN and store training info in history
print()
print('SGD:')
history_SGD = model_DNN_SGD.fit(X_train, Y_train,
          batch_size=batch_size,
          epochs=epochs,
          shuffle=True,
          verbose=1,
          validation_data=(X_test, Y_test))

print()
print('Adam:')
history_Adam = model_DNN_Adam.fit(X_train, Y_train,
          batch_size=batch_size,
          epochs=epochs,
          shuffle=True,
          verbose=1,
          validation_data=(X_test, Y_test))

print()
print('RMSprop:')
history_RMSprop = model_DNN_RMSprop.fit(X_train, Y_train,
          batch_size=batch_size,
          epochs=epochs,
          shuffle=True,
          verbose=1,
          validation_data=(X_test, Y_test))
```

#### Evaluate the Model Performance on the *Unseen* Test Data

```python
# evaluate model
score_SGD = model_DNN_SGD.evaluate(X_test, Y_test, verbose=1)
score_Adam = model_DNN_Adam.evaluate(X_test, Y_test, verbose=1)
score_RMSprop = model_DNN_RMSprop.evaluate(X_test, Y_test, verbose=1)

# print performances
print()
print('SGD')
print('Test loss:', score_SGD[0])
print('Test accuracy:', score_SGD[1])

print()
print('Adam')
print('Test loss:', score_Adam[0])
print('Test accuracy:', score_Adam[1])

print()
print('RMSprop')
print('Test loss:', score_RMSprop[0])
print('Test accuracy:', score_RMSprop[1])
```

```python
def plot_training_history(history, title):
    fig, [ax0, ax1] = plt.subplots(1, 2, figsize=(15,5))
    fig.suptitle(title, size=15)

    # summarize history for accuracy
    ax0.plot(history.history['acc'])
    ax0.plot(history.history['val_acc'])
    ax0.set_ylabel('model accuracy')
    ax0.set_xlabel('epoch')
    ax0.legend(['train', 'test'], loc='best')

    # summarize history for loss
    ax1.plot(history.history['loss'])
    ax1.plot(history.history['val_loss'])
    ax1.set_ylabel('model loss')
    ax1.set_xlabel('epoch')
    ax1.legend(['train', 'test'], loc='best')

    plt.show()

# Plot for each optimizer
plot_training_history(history_SGD, 'DNN with SGD')
plot_training_history(history_Adam, 'DNN with Adam')
plot_training_history(history_RMSprop, 'DNN with RMSprop')
```

Dai grafici si può osservare come il modello con l'ottimizzatore Adam e RMSprop vadano in *overfitting* subito dopo poche epoche, punto in cui si dovrebbe interrompere la sessione di addestramento. Inoltre, l'andamento dell'accuratezza e della funzione di costo è molto seghettata. Ho provato anche a aumentare la grandezza delle *batch*, ma non ho notato grossi miglioramenti. Possiamo concludere che Adam e RMSprop non sono buoni ottimizzatori per questo problema. Come aspettato, all'inizio l'accuratezza del test migliora rispetto a quella di addestramento.

```python
X_test = X_test.reshape(X_test.shape[0], img_rows*img_cols)
predictions_SGD = model_DNN_SGD.predict(X_test)
predictions_Adam = model_DNN_Adam.predict(X_test)
predictions_RMSprop = model_DNN_RMSprop.predict(X_test)

X_test = X_test.reshape(X_test.shape[0], img_rows, img_cols, 1)

plt.figure(figsize=(15, 6)) 
for i in range(20):    
    ax = plt.subplot(2, 10, i + 1)    
    plt.imshow(X_test[i, :, :, 0], cmap='gray')    
    plt.title("Digit: {}\n\nSGD: {}\nAdam: {}\nRMSprop: {}".format(np.argmax(Y_test[i]), np.argmax(predictions_SGD[i]), np.argmax(predictions_Adam[i]), np.argmax(predictions_RMSprop[i])))
    plt.axis('off') 
plt.show()
```

Tutti e tre i modelli riescono a predire correttamente le cifre qui mostrate.

#### Save and Restore the models

```python
save_model_path_SGD='Saved/model_SGD.h5'
save_model_path_Adam='Saved/model_Adam.h5'
save_model_path_RMSprop='Saved/model_RMSprop.h5'
model_DNN_SGD.save(filepath=save_model_path_SGD, include_optimizer=True)
model_DNN_Adam.save(filepath=save_model_path_Adam, include_optimizer=True)
model_DNN_RMSprop.save(filepath=save_model_path_RMSprop, include_optimizer=True)
```

### Exercise 12.2

In questo esercizio viene chiesto di modificare l'architettura della DNN utilizzando strati convoluzionali, approfittando così delle correlazioni spaziali locali tra i *pixel* e dell'invarianza traslazionale presente nelle cifre scritte a mano, per migliorare l'accuratezza del nostro modello di classificazione. Ci sono due tipi di strati di base che compongono una rete neurale convoluzionale (CNN): il *convolutional layer* che calcola la convoluzione dell'*input* con una serie di filtri per estrarre caratteristiche e il *pooling layer* che riduce la risoluzione dell'*input* mantenendo la località e la struttura spaziale. Questi *layer* sono generalmente seguiti da uno strato completamente connesso e da un classificatore come il *soft-max*. 

```python
from keras.layers import Flatten, Conv2D, MaxPooling2D

# reshape data, depending on Keras backend
if keras.backend.image_data_format() == 'channels_first':
    X_train = X_train.reshape(X_train.shape[0], 1, img_rows, img_cols)
    X_test = X_test.reshape(X_test.shape[0], 1, img_rows, img_cols)
    input_shape = (1, img_rows, img_cols)
else:
    X_train = X_train.reshape(X_train.shape[0], img_rows, img_cols, 1)
    X_test = X_test.reshape(X_test.shape[0], img_rows, img_cols, 1)
    input_shape = (img_rows, img_cols, 1)
    
print('X_train shape:', X_train.shape)
print('Y_train shape:', Y_train.shape)
print()
print(X_train.shape[0], 'train samples')
print(X_test.shape[0], 'test samples')
```

```python
def create_CNN():
    # instantiate model
    model = Sequential() #28x28x1
    # add first convolutional layer with 10 filters (dimensionality of output space)
    model.add(Conv2D(10, kernel_size=(5, 5), activation='relu', input_shape=input_shape)) #24x24x10
    # add max pooling layer 
    model.add(MaxPooling2D(pool_size=(2, 2))) #12x12x10
    # add convolutional layer with 20 filters
    model.add(Conv2D(20, kernel_size=(5, 5), activation='relu')) #8x8x20
    # add max pooling layer
    model.add(MaxPooling2D(pool_size=(2, 2))) #4x4x20
    # apply dropout with rate 0.5
    model.add(Dropout(0.5))
    # flat layer
    model.add(Flatten()) #320
    # add a dense tanh layer 
    model.add(Dense(128, activation='tanh'))
    # soft-max layer
    model.add(Dense(num_classes, activation='softmax'))
    # compile the model
    model.compile(loss=keras.losses.categorical_crossentropy, optimizer='SGD', metrics=['acc'])
    return model
```

```python
# training parameters
batch_size = 32
epochs = 50

# create the deep conv net
model_CNN=create_CNN()

# train CNN
history = model_CNN.fit(X_train, Y_train,
          batch_size=batch_size,
          epochs=epochs,
          verbose=1,
          validation_data=(X_test, Y_test))
```

```python
# evaluate model
score = model_CNN.evaluate(X_test, Y_test, verbose=1)

# print performance
print()
print('Test loss:', score[0])
print('Test accuracy:', score[1])

# look into training history

# summarize history for accuracy
plt.plot(history.history['acc'])
plt.plot(history.history['val_acc'])
plt.ylabel('model accuracy')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='best')
plt.show()

# summarize history for loss
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.ylabel('model loss')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='best')
plt.show()
```

Nota che la rete CNN converge più rapidamente rispetto alla rete DNN.

```python
predictions = model_CNN.predict(X_test)

plt.figure(figsize=(15, 5)) 
for i in range(20):    
    ax = plt.subplot(2, 10, i + 1)    
    plt.imshow(X_test[i, :, :, 0], cmap='gray')    
    plt.title("Digit: {}\nPredicted: {}".format(np.argmax(Y_test[i]), np.argmax(predictions[i])))    
    plt.axis('off') 
plt.show()
```

La DNN riesce a predire correttamente tutte le cifre qui mostrate.

```python
save_model_path='Saved/model_CNN.h5'
model_CNN.save(filepath=save_model_path, include_optimizer=True)
```
