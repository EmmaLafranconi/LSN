### Exercise 02.1

In questo esercizio viene chiesto di calcolare il seguente integrale con metodi Monte Carlo,

$$I = \int_0^1 \frac{\pi}{2}\cos(\pi x/2) dx = 1$$

prima campionando una distribuzione uniforme in $[0,1]$ e poi utilizzando l'*importance sampling*.\
Nel primo caso, possiamo scrivere che 

$$I = \lim_{N \to \infty} \frac{1}{N} \sum_{i=1}^N f(x_i) = \langle f \rangle _{[0,1]} $$

e quindi una stima di $I$ può essere ottenuta considerando variabili casuali $x_i$ uniformemente distribuite in $[0,1]$.
L'*importance sampling*, invece, consiste nel campionare variabili casuali non uniformi. Consideriamo una distribuzione di probabilità $p(x)$ (normalizzata) "simile" all'integranda $f(x)$, possiamo scrivere che 

$$ I = \int_0^1 g(x)p(x) dx$$

dove $g(x) = f(x)/p(x)$. In questo caso, una stima di $I$ può essere ottenuta attraverso 

$$ I = \lim_{N \to \infty} \frac{1}{N} \sum_{i=1}^N g(x_i) $$ 

dove $x_i$ sono variabili casuali distribuite come $p(x)$. 

Per risolvere l'integrale con l'*importance sampling*, è stata utilizzata una $p(x) = 3/2(1-x^2)$ e per campionare $p(x)$, è stata utilizzata la tecnica di rigetto.

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import math
```

```python
M=100000000           # Total number of throws
N=100                 # Number of blocks
L=int(M/N)            # Number of throws in each block
x = np.arange(N) + 1  # [0,1,2,...,N-1]
x*=L                  # Number of throws = block * (Number of throws in each block)

int_result = np.loadtxt('int_result.txt', delimiter=' ')
int_IS_result = np.loadtxt('int_IS_result.txt', delimiter=' ')

fig, [ax0, ax1] = plt.subplots(1, 2, figsize=(15,5))

ax0.errorbar(x, int_result[:,0]-1., yerr=int_result[:,1], label='sampling a uniform distribution')
ax0.errorbar(x, int_IS_result[:,0]-1., yerr=int_IS_result[:,1], label='using importance sampling')
ax0.set_xlabel('#throws')
ax0.set_ylabel('I - 1')
ax0.legend()
ax0.grid(True)

ax1.plot(x, int_result[:,1], marker='x', linestyle='none', label='sampling a uniform distribution')
ax1.plot(x, int_IS_result[:,1], marker='.',linestyle='none',  label='using importance sampling')
ax1.set_xlabel('#throws')
ax1.set_ylabel('$\sigma$')
ax1.grid(True)
ax1.legend()

plt.show()
```

Dai grafici risulta evidente come l'applicazione dell'*importance sampling* apporti sostanziali miglioramenti alla stima dell'incertezza e porti a una convergenza più rapida della stima dell'intergale.

### Exercise 02.2

In questo esercizio viene chiesto di simulare un *random walk* 3D su un reticolo cubico e nel continuo. Nella simulazione sono stati considerati 100 passi di lunghezza $a=1$. Per campionare uniformemente una delle tre direzioni principali del reticolo, sono stati generati due numeri casuali: uno per determinare la direzione del passo ($x$, $y$ o $z$) e l'altro per determinare il segno (avanti o indietro). Invece, per campionare uniformemente l'angolo solido, $\theta \in [0,\pi]$ e $\phi \in [0,2\pi]$ sono stati campionati nel seguente modo:

$$ \theta = \cos^{-1} (1-2r) $$
$$ \phi = 2\pi r $$

con $r \in [0,1)$. Per campionare $\theta$ è stato utilizzato il metodo dell'inversione della distribuzione cumulativa. \
Per calcolare la radice del valor medio della distanza dall'origine al quadrato $\sqrt{\langle |\vec{r}_N|^2 \rangle_{RW}}$ in funzione del passo $N \in [0, 100]$, è stato utilizzato il metodo della media a blocchi.


```python
N = np.arange(100) + 1

RW_disc_result = np.loadtxt('RW_disc_result.txt', delimiter=' ')
RW_result = np.loadtxt('RW_result.txt', delimiter=' ')

fig, [ax0, ax1] = plt.subplots(1, 2, figsize=(15,5))

ax0.errorbar(N, RW_disc_result[:,0], yerr=RW_disc_result[:,1], marker='.', label='simulated values', ls='--')
ax0.plot(N, np.sqrt(N), linewidth=4, color='orange', label='expected values')
ax0.set_title('3D Random Walks (RW) on a cubic lattice')
ax0.set_xlabel('#step')
ax0.set_ylabel(r"$\sqrt{\langle |\vec{r}_N|^2\rangle}$")
ax0.grid(True)
ax0.legend()

ax1.errorbar(N, RW_result[:,0], yerr=RW_result[:,1], marker='.', label='simulated values', ls='--')
ax1.plot(N, np.sqrt(N), linewidth=4, color='orange', label='expected values')
ax1.set_title('3D Random Walks (RW) in the continuum')
ax1.set_xlabel('#step')
ax1.set_ylabel(r"$\sqrt{\langle |\vec{r}_N|^2\rangle}$")
ax1.grid(True)
ax1.legend()
plt.show()

plt.figure(figsize=(14,6))
plt.errorbar(N, (RW_disc_result[:,0]-np.sqrt(N))/np.sqrt(N), yerr=RW_disc_result[:,1]/np.sqrt(N))
plt.title('3D Random Walks (RW) on a cubic lattice')
plt.xlabel('#step')
plt.ylabel(r"$(\sqrt{\langle |\vec{r}_N|^2\rangle}-\sqrt{N})~/~\sqrt{N}$")
plt.axhline(y=0, ls='--', c='red')
plt.grid(True)
plt.show()

plt.figure(figsize=(14,6))
plt.errorbar(N, (RW_result[:,0]-np.sqrt(N))/np.sqrt(N), yerr=RW_result[:,1]/np.sqrt(N))
plt.title('3D Random Walks (RW) in the continuum')
plt.xlabel('#step')
plt.ylabel(r"$(\sqrt{\langle |\vec{r}_N|^2\rangle}-\sqrt{N})~/~\sqrt{N}$")
plt.axhline(y=0, ls='--', c='red')
plt.grid(True)
plt.show()
```

Entrambi i *random walk* sono compatibili con una legge del tipo $k\sqrt N$ con $k=1$ entro una deviazione standard. Questo risultato è in accordo con il modello teorico di diffusione, il quale prevede una costante di diffusione pari al quadrato della lunghezza del passo $a$.
