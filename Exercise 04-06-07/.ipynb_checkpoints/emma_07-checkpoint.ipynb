### Exercise 07.2

In questo esercizio viene chiesto di stampare i valori istantanei dell'energia potenziale per particella $U/N$ durante le simulazioni di $M = 5 \times 10^5$ passi Monte Carlo, nelle seguenti condizioni termodinamiche: 

1. fase solida: $\rho^\star = 1.1$, $T^\star = 0.8$ (raggio di cut-off: $r_c = 2.2$)
2. fase liquida: $\rho^\star = 0.8$, $T^\star = 1.1$ (raggio di cut-off: $r_c = 2.5$)
3. fase gassosa: $\rho^\star = 0.05$, $T^\star = 1.2$ (raggio di cut-off: $r_c = 5.0$)

Notiamo che in questo caso non c'è bisogno di partire da una temperatura diversa da quella *target*, poichè essendo in *ensemble* NVT questa è costante. Per ottenere il 50% di accettazione nell'algoritmo di Metropolis, sono stati assunti i seguenti valori per il passo $\delta$ della probabilità di transizione: 

$\delta_{solid} = 0.06$ \
$\delta_{liquid} = 0.1$ \
$\delta_{gas} = 9~(\sim 60 \%)$

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import math
```

```python
equi_penergy_solid = np.loadtxt('OUTPUT/Solid_NVT/equilibration_penergy.dat')
equi_penergy_liquid = np.loadtxt('OUTPUT/Liquid_NVT/equilibration_penergy.dat')
equi_penergy_gas = np.loadtxt('OUTPUT/Gas_NVT/instantaneous_penergy.dat')
```

```python
phase = ['Solid', 'Liquid', 'Gas']
equi_penergy = [equi_penergy_solid[:1000], equi_penergy_liquid[:1000], equi_penergy_gas[:1000]]
step = np.arange(len(equi_penergy_solid[:1000])) + 1

for i in range(0,3):
    plt.figure(figsize=(14,6))
    plt.plot(step, equi_penergy[i])
    plt.title(phase[i]+' Phase')
    plt.xlabel('#steps')
    plt.ylabel(r'$U/N$')
    plt.grid(True)
    plt.show()
```

Dai grafici è evidente come nella simulazione della fase solida e liquida, sia necessario un tempo di equilibrazione di circa 200 passi prima di iniziare con le misure effettive, mentre il gas si trova già all'equilibrio. Fatta l'equilibrazione, calcoliamo la funzione di autocorrelazione temporale di $U/N$ per misurare il tempo di correlazione della simulazione, ovvero la scala temporale tipica entro cui l'autocorrelazione diminuisce, e stimare approssimativamente quanto grandi dovrebbero essere i blocchi nelle simulazioni.

```python
penergy_solid = np.loadtxt('OUTPUT/Solid_NVT/instantaneous_penergy.dat')
penergy_liquid = np.loadtxt('OUTPUT/Liquid_NVT/instantaneous_penergy.dat')
penergy_gas = np.loadtxt('OUTPUT/Gas_NVT/instantaneous_penergy.dat')
```

```python
def autocorrelation(data,taumax):
    chi = np.zeros(taumax)
    tmax = len(data)
    sum4 = np.sum(data**2) 
    sum5 = np.sum(data) 
        
    for t in range(taumax):
        sum1 = np.sum(data[:tmax-t] * data[t:]) #compute the sum of data[i] * data[i+t] for i in [0, tmax-t)
        sum2 = np.sum(data[:tmax-t]) #compute the sum of data[i] for i in [0, tmax-t)
        sum3 = np.sum(data[t:]) #compute the sum of data[i+t] for i in [0, tmax-t)
        
        chi[t] = ((1/(tmax-t)) * sum1 - (1/(tmax-t)) * sum2 * (1/(tmax-t)) * sum3) / ((1/tmax) * sum4 - ((1/tmax) * sum5)**2)
    return chi
```

```python
auto_solid = autocorrelation(penergy_solid, 300)
auto_liquid = autocorrelation(penergy_liquid, 300)
auto_gas = autocorrelation(penergy_gas, 300)
```

```python
t = np.arange(300)
plt.figure(figsize=(14,6))
plt.plot(t, auto_solid, label='Solid Phase')
plt.plot(t, auto_liquid, label='Liquid Phase')
plt.plot(t, auto_gas, label='Gas Phase')
plt.xlabel('time $t$')
plt.ylabel('potential energy autocorrelation $\chi (t)$')
plt.legend()
plt.grid()
plt.show()
```

Come aspettato, l'autocorrelazione decresce in modo esponenziale a tempi lunghi. Per ottenere campioni statisticamente indipendenti, sono stati prelevati ad intervalli superiori al tempo di correlazione. In particolare, sono stati impostati i seguenti valori per numero di blocchi $N$ e lunghezza dei blocchi $L$:

1. fase solida: $N=250$, $L=2000$
2. fase liquida: $N=250$, $L=2000$
3. fase gassosa: $N=100$, $L=5000$

Stimiamo ora le incertezze statistiche del valore di aspettazione dell'energia potenziale per particella $U/N$ con il metodo della media a blocchi, per diverse lunghezze dei blocchi, da $L=10$ a $L=5000$.

```python
def error(av, av2, n):  
    if n==0: return 0
    else: return np.sqrt((av2 - av**2)/n) 
    
def data_blocking(data, L):
    M = len(data) # Total number of throws
    N = M // L      # Number of blocks 
    ave = 0
    av2 = 0
    
    for i in range(N):
        summ = 0
        for j in range(L):
            k = j+i*L
            summ += data[k]
        ave += summ/L
        av2 += (summ/L)**2
    ave /= N
    av2 /= N
    
    return error(ave,av2,N)
```

```python
Lmin = 10    # Min number of throws in each block
Lmax = 5000  # Max number of throws in each block
M = 500000   # Total number of throws
L = np.array([x for x in range(Lmin,Lmax+1) if (M%x == 0)]) #divisors of M

error_solid, error_liquid, error_gas = [], [], []
for i in L:
    error_solid.append(data_blocking(penergy_solid, i))
    error_liquid.append(data_blocking(penergy_liquid, i))
    error_gas.append(data_blocking(penergy_gas, i))
```

```python
plt.figure(figsize=(14,6))
plt.plot(L, error_solid, 'o-', label='Solid Phase')
plt.plot(L, error_liquid, 'o-', label='Liquid Phase')
plt.plot(L, error_gas, 'o-', label='Gas Phase')
plt.xlabel('# throws in each block')
plt.ylabel('statistical uncertainty')
plt.legend()
plt.grid()
plt.show()
```

Quando i blocchi sono sufficientemente grandi, le medie dei blocchi tendono ad essere statisticamente indipendenti. Questo significa che le correlazioni tra i dati all'interno di un blocco sono state eliminate, e il Teorema del Limite Centrale (CLT) può essere applicato. Di conseguenza, le incertezze statistiche possono essere calcolate in modo accurato e affidabile utilizzando la varianza delle medie dei blocchi.

Al contrario, se i blocchi sono troppo piccoli, le medie dei blocchi non sono indipendenti e le stime delle incertezze saranno sottostimate a causa della correlazione tra i dati. Una volta che la lunghezza dei blocchi è sufficientemente grande da essere indipendenti, le incertezze statistiche stimate si stabilizzano e non cambiano significativamente con ulteriori aumenti della lunghezza del blocco. Questo perché si è raggiunto il "vero" errore statistico.

### Exercise 07.3

In questo esercizio viene chiesto di includere nel codice che ci è stato fornito il calcolo della funzione di distribuzione radiale $g(r)$ in funzione della distanza radiale $r$ tra le particelle, con $r \in [0; L/2]$ (dove $L$ rappresenta la dimensione di un lato del box), così definita:

$$ g(r) = \frac{1}{\rho N \Delta V(r)} \Big\langle \sum \sum_{i \not = j = 1}^N \delta \big[|\vec{r}| - |\vec{r_i}-\vec{r_j}|\big] \Big\rangle$$

dove $\Delta V(r) = \frac{4\pi}{3}\big[(r+dr)^3 -r^3\big]$.

Questa funzione fornisce la probabilità di trovare una coppia di atomi a una distanza $r$ l'uno dall'altro, rispetto alla probabilità attesa per una distribuzione completamente casuale alla stessa densità. Nella pratica, quello che è stato fatto durante la simulazione è stato "riempire" un istogramma di 100 bin aumentando il conteggio del bin corrispondente di 2 ogni volta che si trovavano due particelle a una distanza compresa tra $r$ e $r + dr$. Alla fine della simulazione, è stato normalizzato l'istogramma con la quantità $\rho N \Delta V(r)$.

```python
gofr_solid = np.loadtxt('OUTPUT/Solid_NVT/gofr.dat')
gofr_liquid = np.loadtxt('OUTPUT/Liquid_NVT/gofr.dat')
gofr_gas = np.loadtxt('OUTPUT/Gas_NVT/gofr.dat')
```

```python
plt.figure(figsize=(14,6))
plt.errorbar(gofr_solid[:,0], gofr_solid[:,1], yerr=gofr_solid[:,2], label='Solid Phase')
plt.errorbar(gofr_liquid[:,0], gofr_liquid[:,1], yerr=gofr_liquid[:,2], label='Liquid Phase')
plt.errorbar(gofr_gas[:,0], gofr_gas[:,1], yerr=gofr_gas[:,2], label='Gas Phase')
plt.xlabel('r')
plt.ylabel('g(r)')
plt.xlim(0.0,2.25)
plt.legend()
plt.grid()
plt.show()
```

La ragione per cui $g(r)$ è nulla sotto un certo valore di $r$ è dovuta alla forte repulsione tra le particelle a brevi distanze. Il potenziale di Lennard-Jones ha un minimo a una distanza caratteristica $r_{min}$ dove l'attrazione tra le particelle è massima. Questa distanza rappresenta una configurazione energeticamente favorevole, quindi è probabile trovare particelle separate da questa distanza. Questo si riflette in un picco in $g(r)$ a $r\approx r_{min}$, indicando la presenza di un "guscio" di particelle intorno a questa distanza.

Differenze tra le fasi:

1. Nella fase solida le particelle sono disposte in un reticolo cristallino con una struttura altamente ordinata. Questo ordine si manifesta in $g(r)$ come una serie di picchi ben definiti a distanze specifiche che corrispondono ai vari strati del reticolo cristallino. I gusci nella fase solida sono molto definiti perché le particelle sono vincolate a posizioni fisse relative nel reticolo.

2. Nella fase liquida le particelle hanno una certa libertà di movimento, ma tendono ancora ad accumularsi intorno a distanze energeticamente favorevoli. Questo si traduce in picchi in $g(r)$ simili a quelli della fase solida, ma meno pronunciati e più larghi. I gusci sono visibili, ma meno definiti rispetto alla fase solida, indicando una struttura a corto raggio che si perde a lunghe distanze.

3. Nella fase gassosa le particelle si muovono liberamente e le interazioni tra di loro sono deboli e occasionali. Di conseguenza, $g(r)$ è quasi uniforme, indicando una distribuzione quasi casuale delle particelle. Non ci sono gusci ben definiti come nelle altre due fasi.
