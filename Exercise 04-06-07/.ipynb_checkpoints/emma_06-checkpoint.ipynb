### Il Modello di Ising

In Meccanica Statistica, il modello di Ising 1D è esattamente risolvibile. Il modello di Ising è un modello di un magnete che postula un reticolo con un dipolo magnetico o spin su ogni sito. L'Hamiltoniana per un sistema di N spin con interazione tra i primi vicini è data da (si assume $\mu_B = 1$ e $k_B = 1$):

$$
H = 
-J\sum_{i=1}^N s_i s_{i+1}
-\frac{h}{2}\sum_{i=1}^N (s_i + s_{i+1}) 
$$

dove $\{s_i\}_{i=1,N}$ sono variabili di spin che assumono valori $\{-1,1\}$, $h$ è il campo magnetico esterno e $J > 0$ per un modello ferromagnetico, è la tendenza degli spin ad allinearsi parallelamente. Per convenzione, si assumono condizioni periodiche al contorno tali che $s_{N+1} = s_1$, in modo da garantire che tutti gli spin abbiano lo stesso numero di vicini e la stessa geometria locale e quindi che il sistema sia invariante per traslazione.

La funzione di partizione $Z$ per $N$ spin, può essere calcolata esattamente e risulta essere:

$$
Z=\sum_{\{ s_i \}} e^{-\beta H} = \lambda_1^N + \lambda_2^N
$$

dove $\lambda_{1,2} = e^{\beta J} \cosh (\beta h) \pm \left[ e^{2\beta J} \cosh^2 (\beta h) - 2\sinh (2\beta J) \right]^{1/2}$ e dove $\beta=\frac{1}{k_b T} = \frac{1}{T}$. Una volta calcolata la funzione di partizione, l'energia libera di Helmholtz si ottiene facilmente come $A = -k_B T \ln Z$. Tutte le quantità termodinamiche di interesse possono essere dedotte da $A$ o equivalentemente da $Z$.

- **Energia interna** (con $h = 0$):

$$
\frac{U(N,T)}{N} = \frac{\langle H \rangle}{N}
= \frac{1}{N} \sum_{\{ s_i \}} H \frac{\exp \left( -\beta H\right)}{Z}
= -\frac{1}{N}\frac{\partial \ln Z}{\partial \beta}
\underset{h=0}= -J \tanh (\beta J) \Biggl( \frac{1 + \tanh^{N-2} (\beta J)}{1+\tanh^N (\beta J)} \Biggl)
$$

- **Calore specifico** (con $h = 0$):

$$
\frac{C(N,T)}{N} = \frac{1}{N}\frac{\partial U(N,T)}{\partial T} = \frac{1}{N}k_B\beta^2 \left(\langle H^2 \rangle -\langle H \rangle^2 \right) \underset{h=0}= 
k_B (\beta J)^2 \left\{ 1+\tanh^N (\beta J)+(N-1)\tanh^2 (\beta J)+(N-1)\frac{\tanh^{N-2} (\beta J)}{1+\tanh^N (\beta J)}
- N \left[ \frac{\tanh (\beta J) + \tanh^{N-1} (\beta J)}{1+\tanh^N (\beta J)} \right]^2 \right\}
$$

- **Magnetizzazione** (con $h \not= 0$):

$$
\frac{M(N,T,h)}{N} = -\frac{1}{N}\frac{\partial A}{\partial h} = \frac{1}{N}k_B T \frac{\partial \ln Z}{\partial h} = \frac{1}{N}\left\langle \sum_{i=1,N}s_i \right\rangle \\ = \frac{e^{\beta J}\; \sinh(\beta h)}{Z} \; \Bigl[\lambda_1^{N-1}\cdot \Bigl(1+\frac{e^{\beta J}\; \cosh(\beta h)}{\sqrt{e^{2\beta J} \cosh^2 (\beta h) - 2\sinh (2\beta J)}}\Bigr) + \lambda_2^{N-1}\cdot \Bigl(1-\frac{e^{\beta J}\; \cosh(\beta h)}{\sqrt{e^{2\beta J} \cosh^2 (\beta h) - 2\sinh (2\beta J)}}\Bigr)\Bigr]
$$

Nota che con $h = 0$ la magnetizzazione è nulla perché in 1D non possiamo avere una transizione di fase (ferromagnetica) in assenza di un campo magnetico esterno. Nel modello di Ising 1D non c'è magnetizzazione spontanea. 

- **Suscettibilità magnetica** (con $h = 0$):

$$
\chi(N,T) = \frac{\partial M}{\partial h} = \beta \left[ \left\langle \left( \sum_{i=1,N}s_i \right)^2 \right\rangle - \left\langle \sum_{i=1,N}s_i \right\rangle^2 \right] = \\
\underset{h=0}= \beta \left[ \left\langle \left( \sum_{i=1,N}s_i \right)^2 \right\rangle \right] = \beta e^{2\beta J} \Biggl( \frac{1-\tanh^N (\beta J)}{1+\tanh^N (\beta J)} \Biggl)
$$

### Exercise 06.1

In questo esercizio viene chiesto di simulare un modello di Ising 1D con l'algoritmo di Gibbs e con l'algoritmo di Metropolis per campionare il peso statistico di Boltzmann 

$$p(\{ s_i \}) = \frac{e^{-\beta H}}{Z}$$

e di stimare l'energia interna per spin $U/N$, il calore specifico $C/N$, la magnetizzazione per spin $M/N$ e la suscettibilità magnetica $\chi$, in funzione della temperatura $T$ nell'intervallo $[0.5;2.0]$ per un sistema formato da $N=50$ spin con $J=1$.

In questo caso, la probabilità di transizione dell'algoritmo di Metropolis, consiste nel selezionare in modo casuale un singolo spin $s_k$ dal reticolo e invertirlo con probabilità di accettazione:

$$ A(\nu|\mu) = \min\big[1, \frac{p(\nu)}{p(\mu)}\big] = \min\big[1, e^{-\beta(E_\nu-E_\mu)}\big] $$

dove 

$$E_\nu-E_\mu = 2Js_k^\mu \sum_{i~ n.n. to ~k} s_i^\mu$$

Nell'algoritmo di Gibbs invece, noto anche come algoritmo del bagno termico, la mossa proposta è sempre accettata e lo spin $s_k$ viene posto uguale a +1 o -1, indipendentemente dal suo valore attuale, dalla distribuzione condizionale definita dagli altri spin vicini (*heat-bath*):

$$ p(s_k' = \pm 1 | \{ s_j:j \not= k \}) = \frac{1}{1+\exp(\mp 2\beta J \sum_{i~ n.n. to ~k} s_i)}$$

E' stato deciso di iniziare la simulazione partendo dalla temperatura più alta $T=2.0$, temperatura per cui la configurazione iniziale degli spin si trova all'equilibrio, e di ripeterla diminuendo man mano la temperature di 0.25, ogni volta riprendendo dalla configurazione precedente degli spin. In questo modo non è stato necessario aspettare un tempo di equilibrazione prima di iniziare con le misure effettive.

Per evitare *bias* nel calcolo delle incertezze statistiche, poichè esiste una forte correlazione tra configurazioni successive visitate con l'algoritmo di Metropolis (e di Gibbs), è stato utilizzato il metodo della media a blocchi. Infatti, anche se le misurazioni sono correlate, la media nei blocchi è non correlata se i blocchi sono sufficientemente grandi. In particolare sono stati considerati $N=20$ blocchi di lunghezza $L=200000$.

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import math
```

```python
points=100
T = np.linspace(0.2,2.2,num=points)
J = 1.0
Ns = 50
h=0.02 #external field
temp = np.array([2.0, 1.75, 1.5, 1.25, 1.0, 0.75, 0.5])

def energy(T, J, Ns):
    beta = 1/T
    th = np.tanh(J/T)
    thN= th**Ns
    ch = 1/th
    return -J*( th + ch*thN )/( 1 + thN )

def heat(T, J, Ns):
    beta = 1/T
    th = np.tanh(J/T)
    thN= th**Ns
    ch = 1/th
    return ((beta*J)**2)*(((1+thN+(Ns-1)*(th**2)+(Ns-1)*(ch**2)*thN)/(1+thN))-Ns*((th+ch*thN)/(1+thN))**2)

def magnet(T, J, Ns, h):
    b = 1/T
    l1 = np.exp(b*J)*np.cosh(b*h)+np.sqrt(np.exp(2*b*J)*np.cosh(b*h)*np.cosh(b*h)-2*np.sinh(2*b*J))
    l2 = np.exp(b*J)*np.cosh(b*h)-np.sqrt(np.exp(2*b*J)*np.cosh(b*h)*np.cosh(b*h)-2*np.sinh(2*b*J))
    Z = l1**Ns + l2**Ns
    return (np.exp(b*J)*np.sinh(b*h)*((l1**(Ns-1))*(1+np.exp(b*J)*np.cosh(b*h)/np.sqrt(np.exp(2*b*J)*np.cosh(b*h)*np.cosh(b*h)-2*np.sinh(2*b*J))) + (l2**(Ns-1))*(1-np.exp(b*J)*np.cosh(b*h)/np.sqrt(np.exp(2*b*J)*np.cosh(b*h)*np.cosh(b*h)-2*np.sinh(2*b*J)))))/(Z)

def chi(T, J, Ns):
    beta = 1/T
    th = np.tanh(J/T)
    thN= th**Ns
    return beta*np.exp(2*beta*J)*(1-thN)/(1+thN)
```

```python
fig, [ax0, ax1] = plt.subplots(1, 2, figsize=(15,5))
fig.suptitle('Ising 1D, Internal Energy')
energy_MRT2 = np.loadtxt('OUTPUT/Ising_MRT2/total_energy.dat', usecols=(2, 3), skiprows=1)
energy_Gibbs = np.loadtxt('OUTPUT/Ising_Gibbs/total_energy.dat', usecols=(2, 3), skiprows=1)

ax0.plot(T, energy(T,J,Ns), ls='--', label='theoretical model')
ax0.errorbar(temp, energy_MRT2[:,0], yerr=energy_MRT2[:,1], marker='o', linestyle='', label='Metropolis Sampling')
ax0.errorbar(temp, energy_Gibbs[:,0], yerr=energy_Gibbs[:,1], marker='.', linestyle='', label='Gibbs Sampling')
ax0.set_xlabel('T')
ax0.set_ylabel('U/N')
ax0.legend()

ax1.errorbar(temp, (energy_MRT2[:,0]-energy(temp,J,Ns))/energy(temp,J,Ns), yerr=energy_MRT2[:,1]/energy(temp,J,Ns), fmt='o', capsize=5, capthick=1.5, c='orange', label='Metropolis Sampling')
ax1.errorbar(temp, (energy_Gibbs[:,0]-energy(temp,J,Ns))/energy(temp,J,Ns), yerr=energy_Gibbs[:,1]/energy(temp,J,Ns), fmt='o', capsize=5, capthick=1.5, c='green', label='Gibbs Sampling')
ax1.axhline(y=0, ls='--')
ax1.set_xlabel('T')
ax1.set_ylabel('$(U/N - U/N_{model})/U/N_{model}$')
ax1.legend()
plt.show()
```

```python
fig, [ax0, ax1] = plt.subplots(1, 2, figsize=(15,5))
fig.suptitle('Ising 1D, Specific Heat')
cv_MRT2 = np.loadtxt('OUTPUT/Ising_MRT2/specific_heat.dat', usecols=(2, 3), skiprows=1)
cv_Gibbs = np.loadtxt('OUTPUT/Ising_Gibbs/specific_heat.dat', usecols=(2, 3), skiprows=1)

ax0.plot(T, heat(T,J,Ns), ls='--', label='theoretical model')
ax0.errorbar(temp, cv_MRT2[:,0], yerr=cv_MRT2[:,1], marker='o', linestyle='', label='Metropolis Sampling')
ax0.errorbar(temp, cv_Gibbs[:,0], yerr=cv_Gibbs[:,1], marker='.', linestyle='', label='Gibbs Sampling')
ax0.set_xlabel('T')
ax0.set_ylabel('c')
ax0.legend()

ax1.errorbar(temp, (cv_MRT2[:,0]-heat(temp,J,Ns))/heat(temp,J,Ns), yerr=cv_MRT2[:,1]/heat(temp,J,Ns), fmt='o', capsize=5, capthick=1.5, c='orange', label='Metropolis Sampling')
ax1.errorbar(temp, (cv_Gibbs[:,0]-heat(temp,J,Ns))/heat(temp,J,Ns), yerr=cv_Gibbs[:,1]/heat(temp,J,Ns), fmt='o', capsize=5, capthick=1.5, c='green', label='Gibbs Sampling')
ax1.axhline(y=0, ls='--')
ax1.set_xlabel('T')
ax1.set_ylabel('$(c - c_{model})/c_{model}$')
ax1.legend()
plt.show()
```

```python
fig, [ax0, ax1] = plt.subplots(1, 2, figsize=(15,5))
fig.suptitle('Ising 1D, Magnetization (h = 0.02)')
magnet_MRT2 = np.loadtxt('OUTPUT/Ising_MRT2/magnetization.dat', usecols=(2, 3), skiprows=1)
magnet_Gibbs = np.loadtxt('OUTPUT/Ising_Gibbs/magnetization.dat', usecols=(2, 3), skiprows=1)

ax0.plot(T, magnet(T,J,Ns,h), ls='--', label='theoretical model')
ax0.errorbar(temp, magnet_MRT2[:,0], yerr=magnet_MRT2[:,1], marker='o', linestyle='', label='Metropolis Sampling')
ax0.errorbar(temp, magnet_Gibbs[:,0], yerr=magnet_Gibbs[:,1], marker='.', linestyle='', label='Gibbs Sampling')
ax0.set_xlabel('T')
ax0.set_ylabel('$M$')
ax0.legend()

ax1.errorbar(temp, (magnet_MRT2[:,0]-magnet(temp,J,Ns,h))/magnet(temp,J,Ns,h), yerr=magnet_MRT2[:,1]/magnet(temp,J,Ns,h), fmt='o', capsize=5, capthick=1.5, c='orange', label='Metropolis Sampling')
ax1.errorbar(temp, (magnet_Gibbs[:,0]-magnet(temp,J,Ns,h))/magnet(temp,J,Ns,h), yerr=magnet_Gibbs[:,1]/magnet(temp,J,Ns,h), fmt='o', capsize=5, capthick=1.5, c='green', label='Gibbs Sampling')
ax1.axhline(y=0, ls='--')
ax1.set_xlabel('T')
ax1.set_ylabel('$(M - M_{model})/M_{model}$')
ax1.legend()
plt.show()
```

```python
fig, [ax0, ax1] = plt.subplots(1, 2, figsize=(15,5))
fig.suptitle('Ising 1D, Susceptibility')
chi_MRT2 = np.loadtxt('OUTPUT/Ising_MRT2/susceptibility.dat', usecols=(2, 3), skiprows=1)
chi_Gibbs = np.loadtxt('OUTPUT/Ising_Gibbs/susceptibility.dat', usecols=(2, 3), skiprows=1)

ax0.plot(T, chi(T,J,Ns), ls='--', label='theoretical model')
ax0.errorbar(temp, chi_MRT2[:,0], yerr=chi_MRT2[:,1], marker='o', linestyle='', label='Metropolis Sampling')
ax0.errorbar(temp, chi_Gibbs[:,0], yerr=chi_Gibbs[:,1], marker='.', linestyle='', label='Gibbs Sampling')
ax0.set_xlabel('T')
ax0.set_ylabel('$\chi$')
ax0.legend()

ax1.errorbar(temp, (chi_MRT2[:,0]-chi(temp,J,Ns))/chi(temp,J,Ns), yerr=chi_MRT2[:,1]/chi(temp,J,Ns), fmt='o', capsize=5, capthick=1.5, c='orange', label='Metropolis Sampling')
ax1.errorbar(temp, (chi_Gibbs[:,0]-chi(temp,J,Ns))/chi(temp,J,Ns), yerr=chi_Gibbs[:,1]/chi(temp,J,Ns), fmt='o', capsize=5, capthick=1.5, c='green', label='Gibbs Sampling')
ax1.axhline(y=0, ls='--')
ax1.set_xlabel('T')
ax1.set_ylabel('$(\chi - \chi_{model})/\chi_{model}$')
ax1.legend()
plt.show()
```

Dai grafici si nota una buona compatibilità tra le simulazioni e le curve del modello teorico, fatte sia campionando con Metropolis che con Gibbs.
